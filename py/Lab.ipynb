Now let's look at a real life example of linear regression using the <span style="color:lightgreen">Boston housing dataset</span>

This dataset contains information collected by the United States Census Service, concerning housing in the area of Boston Mass. Each instance in this dataset <span style="color:lightgreen">represents a specific town or area</span> within the Boston, Massachusetts region

Here, our independent variable, ie the attribute will be <span style="color:lightgreen">"RM"</span>, which signifies the <span style="color:lightgreen">average number of rooms per dwelling<span>

The dependent variable, ie the numerical metric we will be predicting will be <span style="color:lightgreen">"MEDV"</span>, ie the <span style="color:lightgreen">median value of owner-occupied homes in $1000's</span>

<br><br><br>

Importing all the necessary modules and initialising matplotlib

```python
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import pandas as pd

plt.style.use('dark_background')
# Setting the figure DPI, sort of like quality
plt.rcParams['figure.dpi'] = 100
# Changing the size of markers
plt.rcParams['lines.markersize'] = 2
# Changing the width of line
plt.rcParams['lines.linewidth'] = 1
```

<br><br><br>

Defining our dependent and independent variables

```python
ATTR  = 'RM'
LABEL = 'MEDV'
```

<br><br><br>

Reading the dataset, removing the missing values, and taking a look at the distribution

```python
df = pd.read_csv('HousingData.csv')
df = df[[ATTR, LABEL]]

# Removing all the NA values
df = df.dropna()
# Plotting the distribution
plt.scatter(x=df[ATTR], y=df[LABEL])
# Setting the x and y labels
plt.xlabel(ATTR)
plt.ylabel(LABEL)

plt.show()
```

<br><br><br>

Training a linear regression model and plotting the best fit line

```python
# Initialising the linear regression model
model = LinearRegression()

# Training the model
X = df[[ATTR]]
Y = df[LABEL]
model.fit(X, Y)

# Printing the (approximate) equation of the best fit line
m = round( model.coef_[0]   , 2)
b = round( model.intercept_ , 2)
print(f'y = {m}x + {b}')

# Plotting the best fit line
plt.scatter(x=df[ATTR], y=df[LABEL])
xlim, ylim = plt.xlim(), plt.ylim()
plt.plot(X, model.predict(X), color='red')
plt.xlabel(ATTR)
plt.ylabel(LABEL)
plt.xlim(xlim)
plt.ylim(ylim)

plt.show()
```
